```python
import composite_geometry as cg
```

```python
bv = cg.volumes.CylindricalBuildVolume(20*1e-3, 20*1e-3)
rf = cg.reinforcement.SphereReinforcement(radius=5*1e-3, lattice=cg.tiling.BCCLattice(spacing=12*1e-3))
gmsh_matrix, gmsh_full = cg.meshes.generate_meshes(bv, rf)
```

```python
import plotly.graph_objects as go

msh = gmsh_matrix

# --------------------
# Visualize the geometry
# --------------------
x, y, z = msh.points.T
i, j, k = msh.cells_dict['triangle'].T

fig = go.Figure(data=[
    go.Mesh3d(
        x=x, y=y, z=z,
        i=i, j=j, k=k,
        showscale=True,name='y', 
        opacity=0.5
    )
])

fig.show()
```
